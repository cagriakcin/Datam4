```python
#Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
```

```python
#Hypothesis:

#Null Hypothesis -> There's no significant difference in median value between houses bounded and not bounded by the Charles River

#Alternative Hypothesis -> There's a significant difference in median value between houses bounded and not bounded by the Charles River
```

```python
boston_df.loc[(boston_df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
boston_df.loc[(boston_df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'
boston_df.head(5)
```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS_T'] == 'FAR']['MEDV'], 
                      boston_df[boston_df['CHAS_T'] == 'NEAR']['MEDV'], equal_var = True)
```

```python
#p-value is less than 0.05, we reject the Null Hypothesis, meaning there is not a statistical difference in median value betwenn houses near the Charles River and houses far away
```

```python
#Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)
```

```python
#Hypothesis:

#Null Hypotesis: There isn't statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940

#Alternative Hypothesis: There is statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940

```

```python
from statsmodels.formula.api import ols
lm = ols('MEDV ~ AGE', data = boston_df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

```python
#p-value is less than 0.05, we fail to accept the Null Hypothesis --> There is statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940
```

```python
#Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)
```

```python
#Hypothesis:

#Null Hypothesis: Nitric Oxide concentration is not correlated with the proportion of non-retail business acres per town

#Alternative Hypothesis: Nitric Oxide concentration is correlated with the proportion of non-retail business acres per town
```

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

```python
 #Pearson Coefficient is 0.76365 and p-value less than 0.05, we reject the Null Hypothesis as there is a positive correlation between Nitric oxide concentration and proportion of non-retail business acres per town
```

```python
#What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)
```

```python
#Hypothesis:

#Null Hypothesis: There is no impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes.

# Alternative Hypothesis: There is an impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes.
```

```python
x = boston_df['DIS']
y = boston_df['MEDV']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
predisction = model.predict(x)

model.summary()
```

```python
 #p-value is less than 0.05, we can reject the null hypothesis as there is sufficient evidence that there is an impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes.
```
